```python
import requests, lxml
from bs4 import BeautifulSoup
def main(base_url):
    response = requests.get(base_url)
    soup = BeautifulSoup(response.text, "html.parser")
    print(soup.prettify())
    
main('https://meps.ahrq.gov/data_stats/download_data_files_detail.jsp?cboPufNumber=HC-224')
```
